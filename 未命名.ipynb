```python
table = pd.pivot_table(df, values='等级', columns = '评分人', index = '书名')
```

```python
tp = [("任妍", "韩旭"), ("刘鋆钰", "杨洋"), ("祁婉婷", "葛瑞") ]
res = []

for pair in tp:
    p1,p2 = pair
    temp = table[[p1,p2]]
    temp = temp.dropna #(row = "NaN")
    corr = pd.corr()  #计算pearson相关系数；data.corr('spearman')计算spearman相关系数; data.corr('kendall')计算Kendall相关系数
    res.append(corr)
```
